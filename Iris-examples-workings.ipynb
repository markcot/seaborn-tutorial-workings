# Iris dataset

This is my workings of the iris dataset with seaborn.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")
```

```python
uniform_data = np.random.rand(10, 12)
data = pd.DataFrame({'x':np.arange(1,101),
                     'y':np.random.normal(0,4,100)})

iris = sns.load_dataset("iris")
titanic = sns.load_dataset("titanic")
```

```python
f, ax = plt.subplots(figsize=(5,6))
sns.set()
sns.set_style("whitegrid")
sns.set_style("ticks",
              {"xtick.major.size":8,
               "ytick.major.size":8})
#sns.axes_style("whitegrid")
sns.set_context("notebook",
                font_scale=1.5,
                rc={"lines.linewidth":2.5})
sns.set_palette("husl",3)
sns.color_palette("husl")
flatui = ["#9b59b6","#3498db","#95a5a6","#e74c3c","#34495e","#2ecc71"]
sns.set_palette(flatui)
```

```python
g = sns.FacetGrid(titanic,
                  col="survived",
                  row="sex")
g = g.map(plt.hist,"age")
g.despine(left=True)
g.set_ylabels("Survived")
g.set_xticklabels(rotation=45)
g.set_axis_labels("Survived",
                  "Sex")
```

```python
sns.factorplot(x="pclass",
               y="survived",
               hue="sex",
               data=titanic)
```

```python
sns.lmplot(x="sepal_width",
           y="sepal_length",
           hue="species",
           data=iris)
```

```python
h = sns.PairGrid(iris)
h = h.map(plt.scatter)
h.set(xlim=(0,5),
      ylim=(0,5),
      xticks=[0,2.5,5],
      yticks=[0,2.5,5])
```

```python
sns.pairplot(iris)
```

```python
i = sns.JointGrid(x="x",
                  y="y",
                  data=data)
i = i.plot(sns.regplot,
           sns.distplot)
```

```python
sns.jointplot("sepal_length",
             "sepal_width",
              data=iris,
              kind='kde')
```

```python
sns.stripplot(x="species",
              y="petal_length",
              data=iris)
```

```python
sns.swarmplot(x="species",
              y="petal_length",
              data=iris)
```

```python
sns.barplot(x="sex",
            y="survived",
            hue="class",
            data=titanic)
```

```python
sns.countplot(x="deck",
              data=titanic,
              palette="Greens_d")
```

```python
sns.pointplot(x="class",
              y="survived",
              hue="sex",
              data=titanic,
              palette={"male":"g",
                       "female":"m"},
              markers=["^","o"],
              linestyles=["-","--"])
```

```python
sns.boxplot(x="alive",
            y="age",
            hue="adult_male",
            data=titanic)
```

```python
sns.boxplot(data=iris,orient="h")
```

```python
sns.violinplot(x="age",
               y="sex",
               hue="survived",
               data=titanic)
```

```python
sns.regplot(x="sepal_width",
            y="sepal_length",
            data=iris,
            ax=ax)
```
